### Crime visualizations using Open Data DC

```python
##import packages
import pandas as pd
import numpy as np
import os
from plotnine import *
import scipy.stats as stats 
import warnings
warnings.filterwarnings("ignore")
import re
```

#### 2024 Crime Data

```python
##import data and show the first few rows
crime_2024 = pd.read_csv("../../data/open_data_dc/Crime_Incidents_in_2024_with_neighborhood.csv", encoding='utf-8')
crime_2024.head()
```

```python
##list of neighborhoods of interest
neighborhoods = [
    'Adams Morgan','Anacostia', 'Capitol Hill', 'Columbia Heights', 'Dupont Circle', 
    'H Street', 'Navy Yard', 'NoMa', 'Petworth', 'Shaw', 'U Street', 'Union Station'
]

##join the neighborhoods with the or operator
keywords = '|'.join(neighborhoods)
```

```python
##filter the data to only include the offense and neighborhood name columns and show the first few rows
with_neighborhoods_filter = (crime_2024.filter(["OFFENSE", "Name"]))
with_neighborhoods_filter.head()
```

```python
##filter the data to only show neighborhoods of interest and reset the index
with_neighborhoods_filter_ = with_neighborhoods_filter[with_neighborhoods_filter['Name'].str.contains(keywords, case=False, na=False)]
neighborhoods_cleaned = with_neighborhoods_filter_.reset_index(drop = True)
```

```python
##get the shape of the data
neighborhoods_cleaned.shape
```

```python
##add a count column to the dataframe and show the first few rows
neighborhoods_cleaned["COUNT"] = 1
neighborhoods_cleaned.head()
```

```python
##add a column with the year 
neighborhoods_cleaned["YEAR"] = 2024
```

```python
##Get the total crime by neighborhood by grouping by neighborhood and counting occurences
neighborhoods_cleaned['TotalNeighborhoodCrime'] = neighborhoods_cleaned.groupby('Name')['Name'].transform('count')
```

```python
##filter out the anacostia naval station
neighborhoods_cleaned = neighborhoods_cleaned[~neighborhoods_cleaned['Name'].str.contains('Naval', case=False)]
```

```python
##get the first few rows
neighborhoods_cleaned.head()
```

```python
##Export to a csv
neighborhoods_cleaned.to_csv("../../data/open_data_dc/crime_2024.csv")
```

#### Graph crime data in 2024

```python
##ggplot of crime data by offense in each neighborhood of interest in 2024
crime_2024_plot = (ggplot(data=neighborhoods_cleaned) + 

##bar graph
 geom_bar(

##aesthethics as variable, neighborhood on the x-axis and count on the y with offenses colored in
     aes(x="Name", fill="OFFENSE"), stat="count") +
##minimal theme with labels adjusted to be at a 45 degree angle
    theme_minimal(base_size=8) + 
    theme(axis_text_x=element_text(angle=45, hjust=1)) +


##add labels to both axis, a title and a leged title
labs(x="Neighborhood", y="Crimes Committed", title="Crimes Committed by Neigborhood 2024"))

##show the graph
crime_2024_plot
```

```python
##save the graph as a png
ggsave(crime_2024_plot, "../../data/open_data_dc/crime_2024.png")
```

#### Functions for merging data and plotting crime for 2019-2023

```python
##function to merge crime data by year 
def merge_df(crime_csv, year):
    '''
    clean crime data from open data dc for a specific year
    input: csv for that year, year as an int
    output: cleaned csv for that year
    '''
    crime = pd.read_csv(crime_csv)
    with_neighborhoods_filter = crime[crime['Name'].str.contains(keywords, case=False, na=False)]
    with_name_filter = with_neighborhoods_filter.reset_index(drop = True)
    neighborhoods_cleaned = (with_name_filter.filter(["OFFENSE", "Name"]))
    neighborhoods_cleaned['YEAR'] = year
    neighborhoods_cleaned['TotalNeighborhoodCrime'] = neighborhoods_cleaned.groupby('Name')['Name'].transform('count')
    neighborhoods_cleaned = neighborhoods_cleaned[~neighborhoods_cleaned['Name'].str.contains('Naval', case=False)]
    return neighborhoods_cleaned
```

```python
##function to graph crime category by neighborhood
def crime_cluster(df, x, fill, year):
    '''
    graph the crime data from open data dc for a specific year
    input: crime dataframe, x-axis column, fill data, year as int
    output: bar graph
    '''
    plot = (ggplot(data=df) + 
    geom_bar(
     aes(x=x, fill=fill), stat="count") +
    theme_minimal(base_size=8) + 
    theme(axis_text_x=element_text(angle=45, hjust=1)) +
    labs(x="Neighborhood", y="Crimes Committed", title= "Crimes Committed by Neigborhood " + year))
    return plot
```
