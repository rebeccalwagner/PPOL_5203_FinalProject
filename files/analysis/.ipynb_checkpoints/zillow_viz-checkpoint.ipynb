### Zillow Data

```python
##Setup packages
import pandas as pd
import numpy as np
import os
from plotnine import *
import scipy.stats as stats 
import warnings
warnings.filterwarnings("ignore")
```

#### Clean the housing data from Zillow

```python
##import housing data
housing_data = pd.read_csv("../../data/zillow/zillow_housing_neighborhood_dc.csv", encoding='utf-8')
```

```python
##Get the first few rows of the Zillow data
housing_data.head()
```

```python
##List with neighborhood names for filtering
neighborhoods = [
    'Adams Morgan','Anacostia', 'Capitol Hill', 'Columbia Heights', 'Dupont', 
    'H Street', 'Navy Yard', 'NoMa', 'Petworth', 'Shaw', 'U Street'
]

##Query with neighborhoods joined by or to query the Zillow data
query_string = ' or '.join([f'RegionName == "{neighborhood}"' for neighborhood in neighborhoods])
```

```python
##Filter the Zillow housing data to Region Name, State, City and months from January 2019 until October 2024
housing_data_filtered =  (
    housing_data.filter(['RegionName','State','City', '1/31/19', '2/28/19', '3/31/19', '4/30/19', '5/31/19', '6/30/19',
                        '7/31/19', '8/31/19', '9/30/19', '10/31/19', '11/30/19', '12/31/19','1/31/20', '2/29/20', '3/31/20', 
                         '4/30/20', '5/31/20', '6/30/20', '7/31/20', '8/31/20', '9/30/20', '10/31/20', '11/30/20', '12/31/20',
                         '1/31/21', '2/28/21', '3/31/21', '4/30/21', '5/31/21', '6/30/21', '7/31/21', '8/31/21', '9/30/21', 
                         '10/31/21', '11/30/21', '12/31/21', '1/31/22', '2/28/22', '3/31/22', '4/30/22', '5/31/22', '6/30/22', '7/31/22', 
                         '8/31/22', '9/30/22', '10/31/22', '11/30/22', '12/31/22', '1/31/23', 
                         '2/28/23', '3/31/23', '4/30/23', '5/31/23', '6/30/23', '7/31/23', '8/31/23', '9/30/23', '10/31/23', '11/30/23', 
                         '12/31/23', '1/31/24', '2/29/24', '3/31/24', '4/30/24', '5/31/24', '6/30/24', '7/31/24', '8/31/24', '9/30/24', 
                         '10/31/24']).
    melt(id_vars=['RegionName', 'State', 'City'], var_name='EndMonth', value_name='MedianHomeValue'). ##Melt dates into one column called end month, value is the median home value for that month
    query(query_string) ##use the query to filter the rows to only neighborhoods of interest
                         )
```

```python
##Get the shape of the filtered dataframe
housing_data_filtered.shape
```

```python
##Remove NaN values
housing_data_filtered_na = housing_data_filtered[~housing_data_filtered['MedianHomeValue'].isna()]
##Set the months to date/time format
housing_data_filtered_na['EndMonth'] = pd.to_datetime(housing_data_filtered_na['EndMonth'])
##Check the shape again
housing_data_filtered_na.shape
```

```python
##Export the filtered data to a csv
housing_data_filtered_na.to_csv("../../data/zillow_house_values.csv", index = False)
```

#### Graph the Zillow Home Values by Neighborhood

```python
##ggplot with the Zillow housing data
housing_plot = (ggplot(data=housing_data_filtered_na) + 

##geom smooth for a line graph
 geom_smooth(

##aesthethics as variable, end month on x axis, home value on y and lines colored by region name(neighborhood)
     aes(x="EndMonth", y="MedianHomeValue", color="RegionName"), method='loess', se=False) +
##theme with a size of 8
    theme_minimal(base_size=8) + 
 


##label the x and y axis and add a title to the graph and the legend
labs(x="Month and Year", y="Median Home Value", title="Median Home Value 2019-2024 by Neighborhood", color="Neighborhood"))

##Show the graph
housing_plot
```

```python
##save the plot as a png
ggsave(housing_plot, "../../data/zillow/housing_values.png")
```

#### Clean the rental data from Zillow 

```python
##import Zillow rental data
rental_data = pd.read_csv("../../data/zillow/zillow_rental_value_zipcode.csv", encoding='utf-8')
```

```python
##import zip code to neighborhood file and show the first few rows
zip_code = pd.read_csv("../../data/geospacial/geo_matching.csv")
zip_code.head()
```

```python
##show the first few rows of the rental data
rental_data.head()
```

```python
##Filter the Zillow rental data to Region Name, State, City and months from January 2019 until October 2024
rental_data_filter = (rental_data.filter(['RegionName','State','City', '2019-01-31', '2019-02-28', '2019-03-31', 
                                          '2019-04-30', '2019-05-31', '2019-06-30', '2019-07-31', '2019-08-31', 
                                          '2019-09-30', '2019-10-31', '2019-11-30', '2019-12-31', '2020-01-31', 
                                          '2020-02-29', '2020-03-31', '2020-04-30', '2020-05-31', '2020-06-30', 
                                          '2020-07-31', '2020-08-31', '2020-09-30', '2020-10-31', '2020-11-30', 
                                          '2020-12-31','2021-01-31', '2021-02-28', '2021-03-31', 
                                          '2021-04-30', '2021-05-31', '2021-06-30', '2021-07-31', '2021-08-31', 
                                          '2021-09-30', '2021-10-31', '2021-11-30', '2021-12-31', '2022-01-31', 
                                          '2022-02-28', '2022-03-31', '2022-04-30', '2022-05-31', '2022-06-30', '2022-07-31', 
                                          '2022-08-31', '2022-09-30', '2022-10-31', '2022-11-30', '2022-12-31', 
                                          '2023-01-31', '2023-02-28', '2023-03-31', '2023-04-30', '2023-05-31', '2023-06-30', 
                                          '2023-07-31', '2023-08-31', '2023-09-30', '2023-10-31', '2023-11-30', '2023-12-31',
                                          '2024-01-31', '2024-02-29', '2024-03-31', '2024-04-30', '2024-05-31', '2024-06-30', 
                                          '2021-07-31', '2021-08-31', '2024-09-30', '2024-10-31']).
                      melt(id_vars=['RegionName', 'State', 'City'], var_name='EndMonth', value_name='MedianRentalPrice'). ##Melt dates into one column called end month, value is the median rental price for that month
                      query('State == "DC"') ##limit the state to DC
                     )
```

```python
##Get the first few rows of the filtered dataframe
rental_data_filter.head()
```

```python
##Merge the filtered data with the zip code to neighborhood data on zip code
rental_data_neighborhood = pd.merge(rental_data_filter, zip_code, how = "left", left_on = "RegionName", right_on = "Zip")
##show the last few rows
rental_data_neighborhood.tail()
```

```python
##List of neighborhoods to search for
neighborhoods = [
    'Adams Morgan','Historic Anacostia', 'Capitol Hill', 'Columbia Heights', 'Dupont Circle', 
    'H Street', 'Navy Yard', 'NoMa', 'Petworth', 'Shaw', 'U Street'
]

##join the list together with the or operator
keywords = '|'.join(neighborhoods)
```

```python
##Loop through the data and keep only neighborhoods in the list
rental_data_filter_zip = rental_data_neighborhood[rental_data_neighborhood['Neighborhood'].str.contains(keywords, case=False, na=False)]
neighborhoods_filtered = rental_data_filter_zip.reset_index(drop = True)
##show the last few rows
neighborhoods_filtered.tail()
```

```python
##Clean the dataframe to remove NaN, set the months to date/time format reset the index and show the first few rows
rental_data_filter_na = neighborhoods_filtered[~neighborhoods_filtered['MedianRentalPrice'].isna()]
rental_data_filter_na['EndMonth'] = pd.to_datetime(neighborhoods_filtered['EndMonth'])
cleaned_rental_data = rental_data_filter_na.reset_index(drop = True)
cleaned_rental_data.head()
```

#### Plot the Zillow Rental prices by Neighborhood

```python
##ggplot with the Zillow rental data
rental_plot_neighborhood = (ggplot(data=cleaned_rental_data) + 

##geom smooth for a line graph
 geom_smooth(

##aesthethics as variable, end month on x axis, rental price on y and lines colored by neighborhood
     aes(x="EndMonth", y="MedianRentalPrice", color="Neighborhood"), method='loess', se=False, size = 1) +
##theme with a size of 8
    theme_minimal(base_size=8) + 



##label the x and y axis and add a title to the graph and the legend
labs(x="Month and Year", y="Median Rental Price", title="Median Rental Price 2019-2024 by Neighborhood", color="Neighborhoods"))

##Show the graph
rental_plot_neighborhood
```

```python
##save the graph to a png
ggsave(rental_plot_neighborhood, "../../data/zillow/rental_values_neighborhood.png")
```
